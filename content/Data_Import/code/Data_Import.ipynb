```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Mangul-Lab-USC/CXPT-501-Fall_23/main/Eloise_Liu.csv')
df
```

```python
cleaned_df = df.dropna() # Remove rows with NaN values
cleaned_df
```

```python
cleaned_df.isnull().sum() #Confirm the number of NaN value in each row
```

```python
remaining_patient_data = df[
    (df['Age'] >= 18) &
    (df['Pre-treatment NT-proBNP (pg/mL)'] >= 1600) &
    (df['Pre-treatment BNP (pg/mL)'] >= 400) &
    (df['Left ventricular ejection fraction (%)'] <= 40)
]
remaining_patient_data
```

```python
len(remaining_patient_data)
```

```python
remaining_patient_data['BMI (kg/m^2)'] = remaining_patient_data['Weight (kg)'] / ((remaining_patient_data['Height (cm)'] / 100) ** 2)
remaining_patient_data
```

```python
remaining_patient_data.to_csv('remaining_patient_data.csv', index=False)
```
