##**Unix commands for manipulating files**

Files can be managed with the following commands, using the name of the file as a command line argument.

###**Create an empty file with `cat >` and verify its creation with `ls`.**

Here, `>` is known as a redirection operator. You can think of this syntax as the output of the `cat` command being directed towards some target. Here, we will use it to create our target, an empty file called `emptyFile.txt`.

image.png

```python
%%bash
cat > emptyFile.txt
```

You can confirm that this new file will persist between code blocks with `ls`.

```python
%%bash
ls
```

###**Write to a file with `cat >` and verify its creation with `ls`.**

In addition to writing empty files, you can also add text to a file by writing text directly below `cat > FILENAME`.

```python
%%bash
cat > file.txt
This is the first line of this file.
This is line two of this file.
This is the final line of this file.
```

image.png

Note that rerunning this code block again with different text will overwrite the contents of the existing file.

```python
%%bash
cat > file.txt
This is Line 1.
This is Line 2.
This is Line 3.
```

```python
%%bash
ls file.txt
```

image.png

Not only was the file `file.txt` overwritten, but notice how due to the special nature of how `cat > FILENAME` works in Colab, the command `ls` was not run. Only commands that are placed before `cat > FILENAME` will function normally.

###**Read the entirety of a file at once with `cat`.**

In addition to reading files from Colab's GUI menu. You can also read files from the command line.

```python
%%bash
cat file.txt
```

Since this prints out all the text in a file, be careful not to use this with files that contain hundreds or millions of lines. To read large files, use the `head` command instead of `cat`.

###**Rename `emptyFile.txt` to `empty.txt` with the `mv` command and verify the change with `ls`.**

Doing this takes two command line arguments: first the name of the file to rename followed by what to rename it to.

image.png

```python
%%bash
mv emptyFile.txt empty.txt
ls
```

###**Make a copy of `empty.txt` called `empty2.txt` with the `cp` command and verify its creation with `ls`.**

Doing this also takes two command line arguments: first the name of the file to copy followed by what will be the name of the duplicate file.

```python
%%bash
cp empty.txt empty2.txt
ls
```

###**Remove `empty.txt`, `empty2.txt`, and `file.txt` with the `rm` command then verify their deletion with `ls`.**

While both files can be removed individually, it is faster to remove both of them at once though use of additional command line arguments.

```python
%%bash
rm empty.txt empty2.txt file.txt
ls
```

Similarly, multiple command line arguments can be used to manipulate several files at once with the other Unix commands too. Experiment with this at your leisure!

##**Unix commands for manipulating directories.**

In addition to file mangement, Unix commands can also be used to manage  directories.

###**Use the `mkdir` command to made a subdirectory called newDirectory and verify its creation with `ls`.**

```python
%%bash
mkdir newDirectory
ls
```

You can also create multiple directories at once by providing the command `mkdir` with multiple arguments.

###**Rename the subdirectory `newDirectory` to `projectDirectory` with the `mv` command and verify the change with `ls`.**

Like before, this takes two command line arguments: first the name of the directory to rename followed by what to rename it to.

```python
%%bash
mv newDirectory projectDirectory
ls
```

###**Make a copy of `projectDirectory` called `projectDirectoryTwo` with the `cp` command and verify its creation with `ls`.**

`cp` also takes two command line arguments: first the name of the directory to copy followed by what will be the name of the duplicate directory.



```python
%%bash
cp projectDirectory projectDirectoryTwo
ls
```

Notice that the `cp` command failed. This is due to a safety feature built into both the `cp` and `rm` commands to prevent the accidental duplication or deletion of a directory and all its contents, including any subdirectories. After all, it could be a mess accidentally deleting an entire project or filling up a hard drive with duplicate files.

To bypass this, add the text `-r` after `cp` to enable the copying of directories.

```python
%%bash
cp -r projectDirectory projectDirectoryTwo
ls
```

It worked! Additionally, `-r` here is the abbreviation of the word 'recursive' which describes the copying of a directory, and any subdirectories, sub-subdirectories, sub-sub-subdirectories, etc. As such, you can also type `--recursive` instead.

Such additions that start with either `-` or `--` are called command line options.

image.png

```python
%%bash
cp --recursive projectDirectoryTwo projectDirectoryThree
ls
```

###**Use the `rm` command with the `-r` option to remove `projectDirectory`, `projectDirectoryTwo`, and `projectDirectoryThree` then verify their deletion with `ls`.**

Like with files, multiple directories can also be deleted at once though use of additional command line arguments.

```python
%%bash
rm -r projectDirectory projectDirectoryTwo projectDirectoryThree
ls
```
