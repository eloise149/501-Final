## **What is Google Colab?**

Google Colab is a free cloud-based service provided by Google that allows you to create, edit, and share Jupyter Notebooks. It is designed to make data analysis and machine learning more accessible by providing a platform that requires no setup and runs entirely in the cloud.

## **What is a Jupyter Notebook?**

jupyter.png

A Jupyter Notebook is an interactive document that allows you to combine text, code, and multimedia content in a single document. Jupyter Notebooks are widely used for data analysis, scientific computing, and machine learning because they provide an easy-to-use environment for experimenting with code, documenting findings, and sharing results.

## **Why use Google Colab?**

Google Colab offers several benefits, including:
* *No installation required*: Google Colab runs in your browser, so you can use it on a computer or almost any smart device without needing to install any software.
* *Collaboration*: You can easily share and collaborate on notebooks with others. â˜†
* *Free access to GPUs and TPUs*: Google Colab provides free access to powerful hardware accelerators, making it suitable for machine learning tasks.
* *Integration with Google Drive*: Google Colab can access files stored in your Google Drive, making it easy to work with your data.

## **Runtime type (Python3 or R)**


Google Colab currently supports Python 3 and R as runtime types. By default, a new notebook will use Python 3. Python 3 is the most widely used runtime for data analysis and machine learning, while R is popular for statistical computing and data visualization. â˜†


If you would like to exclusively use R, create an R runtime notebook via this link: https://colab.research.google.com/#create=true&language=r

Running R code natively in an R runtime notebook is faster than using `rpy2` package in a Python3 notebook, especially when it is necessary to install R packages from CRAN, Bioconductor, or GitHub.

## **Connecting Colab to your Google Drive**
To access files stored in your Google Drive from a Google Colab notebook, you need either generate a direct link (see: https://sites.google.com/site/gdocs2direct/ ) or mount your Google Drive as a filesystem. Here, we will demonstrate the latter method.

When you run the following code, you will be prompted to click on a link and authorize the connection with your Google Account. Once you have authorized the connection, copy the authorization code provided and paste it into the input field in the Colab notebook.



```python
from google.colab import drive

drive.mount('/content/google')
```

Now, your Google Drive is mounted as a filesystem, and you can access your files using the /content/gdrive/MyDrive path. You can read and write files directly from your Google Drive, making it easy to work with persistent data and save your progress.

## **Troubleshooting: Resetting runtime**

If you encounter any issues with your notebook, such as a frozen code cell or unexpected errors, you can try deleting the runtime. Deleting the runtime will restart your notebook, clearing all variables and deleting any generated or manually uploaded files while keeping your code and text blocks.







To delete the rutime, click on "Runtime" in the toolbar then "Disconnect and delete runtime". You will be prompted to confirm your action, and once you do, your notebook will restart. While you will also see an option to `Restart runtime`, deleting the runtime is usually preferred to resetting your runtime to ensure the your code runs correctly in a fresh notebook.

## **Troubleshooting: Timeout & instance limitations**


Google Colab has some limitations to prevent abuse:

- **Idle timeout**: If your notebook is idle for more than 90 minutes, your instance will be terminated. To avoid losing work, make sure to save your progress regularly. If you need to run a long computation, make sure to interact with the notebook periodically to prevent it from being considered idle.
- **Maximum runtime**: Google Colab instances have a maximum runtime of 12 hours. After 12 hours, your instance will be terminated, and you will need to restart your notebook.
