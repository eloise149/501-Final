##**Iteration (`for`)**

In Python, for loops start with the keyword `for` followed by a temporary name/variable, the `in` keyword, and lastly an iterable data type. Additionally, the code within each for loop is indented with a single tab per `for`.

###Use a for loop to iterate over characters of a string.

```python
string = 'AGTCAGCA'

for nucleotide in string:
  print(nucleotide)
```

###Use a for loop to iterate over elements of a list.

```python
stopCodonList = ['TAG', 'TAA', 'TGA']

for codon in stopCodonList:
  print(codon)
```

###Use a for loop to iterate over key-value pairs of a dictionary.

```python
aaDict = { 'Met':'AUG', 'Leu': ['UUA', 'UUG', 'CUU', 'CUC', 'CUA', 'CUG'],
           'Ser': ['UCU', 'UCC', 'UCA', 'UCG', 'AGU', 'AGC']}

for aa in aaDict:
  print(aa)
```

###Use a for loop to iterate over a sequence of integers generated by `range()`.

```python
for number in range(0,10):
  print(number)
```

###Use a for loop to iterate over columns of a DataFrame.

```python
import pandas as pd
columns = {'columnOne': ['cell1','cell2','cell3','cell4'], 'columnTwo': [1,2,float('nan'),4]}
df = pd.DataFrame(columns)
df
```

```python
for column in df:
  print(df[column])
```

###Use a for loop to iterate over rows of a DataFrame.

```python
numberOfRows = len(df)

for rowIndex in range(0, numberOfRows):
  print(df.iloc[rowIndex])
```

###Use a nested for loop to iterate over columns then rows of a DataFrame.

```python
for column in df.columns:
  for cell in df[column]:
    print(cell)
```

###Use a nested for loop to iterate over rows then columns of a DataFrame.

```python
numberOfRows = len(df)
numberOfColumns = len(df.columns)

for rowIndex in range(0, numberOfRows):
  for colIndex in range(0, numberOfColumns):
    print(df.iloc[rowIndex, colIndex])
```

##**Condtional statements (`if/else`)**

In Python, if/else statements start with the keyword `if` followed by a condition such as `x is y`. Additional conditions may be specified with the `elif` keyword followed by another condition. The final section of an `if/else` statement is denoted by the keyword `else`.

Additionally, like for loops, the code within each for condition is indented with a single tab.

###Use a simple `if/else` statement.

```python
statement = True

if statement is True:
  print("The statement is true.")
else:
  print("The statement is not true.")
```

```python
statement = False

if statement is True:
  print("The statement is true.")
else:
  print("The statement is not true.")
```

###Use an `if/else` statement with three or more conditions.

```python
sequence = 'CCGGG'

if 'A' in sequence:
  print("The sequence has an Adenine!")
elif 'T' in sequence:
  print("The sequence has an Thymine!")
elif 'C' in sequence:
  print("The sequence has an Cytosine!")
elif 'G' in sequence:
  print("The sequence has an Guanine!")
else:
  print("There are no standard DNA nucleotides in the sequence.")
```

##**Write the code.**

```python
import random

# Define the study and skip categories.
onePercentDose = []
twoPercentDose = []
placebo = []
skip = []

# Group the study categories in a list.
categories = [onePercentDose, twoPercentDose, placebo, skip]

# Define the participants who have not yet been random assigned to a category.
participants = ['Evette', 'Saso', 'Yahweh', 'Oghenekaro','Adhamh',
                'Petronella', 'Kurt', 'Cornelia','Sasori', 'Viola',
                'Candace', 'Revazi', 'Rasul', 'Henny', 'Eugene',
                'Simon', 'Rylie', 'Max', 'Xiong', 'Celina']

# Count how many participants there are to assign.
numberOfParticipantsToAssign = len(participants)

# Loop as many times as there are participants to assign.
# Loop Start ----------------------------------------------------------------|
for loop in range(0, numberOfParticipantsToAssign):

  # Generate a random index of a participant.
  participantIndex = random.randint(0, len(participants)-1)

  # Get a participant from a participant index.
  participant = participants[participantIndex]

  # Generate a random index for a study or skip category.
  categoryIndex = random.randint(0, len(categories)-1)

  # Get a category from a category index.
  category = categories[categoryIndex]

  # Randomly assign a participant to a category.
  category.append(participant)

  # Remove the participant from the list of participants not assigned.
  participants.pop(participantIndex)

# Loop End -----------------------------------------------------------------|

# Print out all the categories and their members.
print('1% THC Dose: ' + str(onePercentDose))
print('2% THC Dose: ' + str(twoPercentDose))
print('Placebo: ' + str(placebo))
print('Skip: ' + str(skip))
```

##**Integrating visualizations**

In addition to performing calculations, statistical programming can also integrate data visualization.

```python
import matplotlib.pyplot as plt
import random

# Define the number of coin tosses.
numberOfTosses = 500

# Create a list containing intergers from 0 to the number of coin tosses.
listOfTossNs = range(numberOfTosses)

# Create a list to store the coin flip results.
flips = []

# Create a list to store the percentage of heads after each coin toss.
listOfPercentHeads = []

# Loop as many times as there are coin tosses.
# Loop Start ----------------------------------------------------------------|
for toss in listOfTossNs:

  # Toss an imaginary coin and record the result.
  flip = random.randint(0, 1)

  # Add the result to the list of coin flip results.
  flips.append(flip)

  # Calculate the proportion of heads.
  percentOfHeads = sum(flips) / len(flips)

  # Add the new proportion to the list storing percent heads.
  listOfPercentHeads.append(percentOfHeads)

# Loop End ------------------------------------------------------------------|

# Plot the graph using Matplotlib.
plt.plot(listOfTossNs, listOfPercentHeads)
# Label the x-axis
plt.xlabel("Number of Tosses")
# Label the y-axis
plt.ylabel("Proportion of Heads")

# Add a horizontal solid line '_' at y = 0.5
# to create a cutoff for P(heads) = 0.5, color that line red
plt.axhline(y=0.5, color='r', linestyle='-')
plt.savefig('test.pdf')
```

##**Creating keyword arguments**

```python
def generateCoinFlipGraph(n):
  """
  Generate a coin flip graph displaying the proportion of heads over time.

  Args:
    n (int): the number of times to flip the coin
  """
  import matplotlib.pyplot as plt
  import random

  # Define the number of coin tosses.
  numberOfTosses = n

  # Create a list containing intergers from 0 to the number of coin tosses.
  listOfTossNs = range(numberOfTosses)

  # Create a list to store the coin flip results.
  flips = []

  # Create a list to store the percentage of heads after each coin toss.
  listOfPercentHeads = []

  # Loop as many times as there are coin tosses.
  # Loop Start ----------------------------------------------------------------|
  for toss in listOfTossNs:

    # Toss an imaginary coin and record the result.
    flip = random.randint(0, 1)

    # Add the result to the list of coin flip results.
    flips.append(flip)

    # Calculate the proportion of heads.
    percentOfHeads = sum(flips) / len(flips)

    # Add the new proportion to the list storing percent heads.
    listOfPercentHeads.append(percentOfHeads)

  # Loop End ------------------------------------------------------------------|

  # Plot the graph using Matplotlib.
  plt.plot(listOfTossNs, listOfPercentHeads)
  # Label the x-axis
  plt.xlabel("Number of Tosses")
  # Label the y-axis
  plt.ylabel("Proportion of Heads")

  # Add a horizontal solid line '_' at y = 0.5
  # to create a cutoff for P(heads) = 0.5, color that line red
  plt.axhline(y=0.5, color='r', linestyle='-')
  plt.savefig('test.pdf')




generateCoinFlipGraph(n = 10)
```

Finally, convert your code into a function that returns a probability matrix given a count matrix. Then, run your function with a single line of code.

Your function also needs to contain a comment which states what the function does and what its arguments are to facilitate its reuse.

```python
def convert_to_probability(countMatrix):
  """
    Converts a count matrix with 'All' margins into a probability matrix.

    Arguments:
  countMatrix (pd.DataFrame): A count matrix with 'All' margins.
  """
  prob_matrix = countMatrix.copy()
  total_patients = countMatrix.iloc[len(countMatrix.index) - 1, len(countMatrix) - 1]
  prob_matrix = prob_matrix.iloc[range(len(prob_matrix.index)), range(len(prob_matrix))] / total_patients
  return prob_matrix

df = pd.read_csv('https://drive.google.com/uc?export=download&id=1L4aScG_IYBOnCCgtjJ1yLKUhi6n68YGP')
countMatrix = pd.crosstab(df['DIABETES_AND_HYPERTEN'], df['CVD'], margins=True)
print(convert_to_probability(countMatrix))
```
